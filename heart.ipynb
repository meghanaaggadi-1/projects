```python
from scipy.stats import norm
import numpy as np
plt.hist(product["age"],bins=20, rwidth=0.8, density=True)
plt.xlabel('Age')
plt.ylabel('Count')

rng = np.arange(product["age"].min(), product["age"].max(), 0.1)
plt.plot(rng, norm.pdf(rng,product["age"].mean(),product["age"].std()))
```

```python
!pip install transformers
from transformers import pipeline
```

```python
# download model
qa_model = pipeline("question-answering")
```

```python
# test model
question = "What is temporal difference?"
question1="where we use temporal difference learning"
question2 = "the value state is based on?"
question3 = "discounted by a factor called?"
question4 = "What is discount rate?"
question5= "What is full form of TD ?"
context = "Temporal difference (TD) learning is a unsupervised learning method used in reinforcement learning to estimate the value of future rewards based on past rewards. In TD learning, the value of a state is based on the expected future reward from that state, discounted by a factor called the discount rate. Discount Rate : The discount rate determines how much importance is placed on future rewards relative to present rewards. Higher discount rates mean that future rewards are valued more highly, while lower discount rates mean that present rewards are valued more highly"
qa_response = qa_model(question = question, context = context)
qa_response1 = qa_model(question = question1, context = context)
qa_response2= qa_model(question = question2, context = context)
qa_response3= qa_model(question = question3, context = context)
qa_response4= qa_model(question = question4, context = context)
qa_response5= qa_model(question = question5, context = context)
print(qa_response)
print(qa_response1)
print(qa_response2)
print(qa_response3)
print(qa_response4)
print(qa_response5)
```
